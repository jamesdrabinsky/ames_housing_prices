## Plotting and Applying Log Transformation to SalePrice

```python
plt.hist(train_df["saleprice"], bins =20, color='b')
plt.show()
```

This histogram has a positive skew and a long tail.  I am going to apply a log transformation and see whether the SalePrice takes on a more normal distribution.

```python
train_df['saleprice'] = np.log(train_df['saleprice'])
```

```python
plt.hist(train_df["saleprice"], bins =20, color='b')
plt.show()
```

This plot looks more like a normal distribution.  One of the assumption of linear regression is distribution normality so I will keep this log transformed SalePrice. 

##  Correlation

Below is a heatmap showing the correlation between the numeric features and SalePrice.

```python
numeric_columns = train_df[train_df.select_dtypes(exclude = object).columns]
corr = numeric_columns.corr()

mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

f, ax = plt.subplots(figsize=(11, 9))
cmap = sns.diverging_palette(220, 10, as_cmap=True)
sns.heatmap(corr, mask=mask, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

While this heatmap is a helpful visual I would like to find the exact correlation between each of the variables.  I will create a matrix showing the correlation between each pair of numeric features and which highlights each cell which has a correlation large than 0.5.
